1. Name any five plots that we can plot using the seaborn library . Also , state the uses of each plot .

1. Scatter Plot
* Use: Scatter plots are used to visualize the relationship between two continuous variables. They are useful for identifying patterns, clusters, or correlations in the data.
* Seaborn Function: sns.scatterplot()

2. Histogram
* Use: Histograms are used to visualize the distribution of a single . continuous variable. They help in understanding the central tendency, spread, and shape of the data.
* Seaborn Function: sns.histplot()

3. Bar Plot
* Use: Bar plots are used to visualize the distribution of a categorical variable or to compare the means of different categories. They are effective for showing relative sizes or frequencies.
* Seaborn Function: sns.barplot() or sns.countplot() for count plots.
4. Box Plot
* Use: Box plots (or box-and-whisker plots) are used to visualize the distribution of a continuous variable across different levels of one or more categorical variables. They provide a summary of the central tendency, spread, and skewness of the data.
* Seaborn Function: sns.boxplot()
5. Heatmap
* Use: Heatmaps are used to visualize the matrix-like data where the values are represented as colors. They are useful for exploring correlations, patterns, or trends in large datasets.
* Seaborn Function: sns.heatmap()

2. Load the "fmri" dataset using the load_dataset function of seaborn . Plot a line plot using x= 'timepoint' and y= 'signal' for different events and regions .


```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "fmri" dataset
fmri_data = sns.load_dataset("fmri")

# Plot a line plot
plt.figure(figsize=(10, 6))
sns.lineplot(data=fmri_data, x='timepoint', y='signal', hue='event', style='region', markers=True, dashes=False)
plt.title('FMRI Signal by Timepoint for Different Events and Regions')
plt.xlabel('Timepoint')
plt.ylabel('Signal')
plt.legend(title='Event')
plt.show()
```

3. Load the "titanic" dataset using the load_dataset function of seaborn . Plot two box plots using x= 'pclass'  , y= 'age'  and y= 'fare' .



```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "titanic" dataset
titanic_data = sns.load_dataset("titanic")

# Plot two box plots
plt.figure(figsize=(12, 6))

# Box plot for 'age' by 'pclass'
plt.subplot(1, 2, 1)
sns.boxplot(data=titanic_data, x='pclass', y='age')
plt.title('Box Plot of Age by Pclass')
plt.xlabel('Pclass')
plt.ylabel('Age')

# Box plot for 'fare' by 'pclass'
plt.subplot(1, 2, 2)
sns.boxplot(data=titanic_data, x='pclass', y='fare')
plt.title('Box Plot of Fare by Pclass')
plt.xlabel('Pclass')
plt.ylabel('Fare')

plt.tight_layout()
plt.show()
```

4. Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column . Use the hue parameter for the 'cut' column of the diamonds dataset .



```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "diamonds" dataset
diamonds_data = sns.load_dataset("diamonds")

# Plot histogram with hue for 'cut' column
plt.figure(figsize=(10, 6))
sns.histplot(data=diamonds_data, x='price', hue='cut', multiple='stack', bins=30)
plt.title('Histogram of Diamond Prices by Cut')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.legend(title='Cut')
plt.show()
```

5.Use the "iris" dataset from seaborn to plot a pair plot . Use the hue parameter for the "species" column of the iris dataset .




```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "iris" dataset
iris_data = sns.load_dataset("iris")

# Plot pair plot with hue for 'species' column
sns.pairplot(iris_data, hue="species")
plt.title('Pair Plot of Iris Dataset with Species Hue')
plt.show()
```

6. Use the "flights" dataset from seaborn to plot a heatmap .

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the flights dataset from seaborn
flights_data = sns.load_dataset("flights")

# Pivot the data to have months as rows and years as columns
flights_pivot = flights_data.pivot_table(index='month', columns='year', values='passengers')

# Create a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(flights_pivot, cmap="YlGnBu", annot=True, fmt="d", linewidths=.5)
plt.title('Passenger counts by Month and Year')
plt.xlabel('Year')
plt.ylabel('Month')
plt.show()
```
