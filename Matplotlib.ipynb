1. What is Matplotlib ? Why is it used ? Name five plots that can be plotted using the Pyplot module of Matplotlib .

Matplotlib is a comprehensive library for creating static, animated, and interactive visualizations in Python. It is widely used for data visualization due to its flexibility and ease of use. Matplotlib allows users to create a wide variety of plots, charts, and graphs to explore and communicate data effectively.

Matplotlib is used for various purposes, including:

1. Data Exploration: Matplotlib enables users to visualize data to gain insights and understand patterns and trends within datasets.
2. Data Communication: It is used to communicate information effectively through visual representations such as plots and charts, making complex data more accessible to a wider audience.
3. Publication-Quality Figures: Matplotlib provides tools to create high-quality figures suitable for publication in scientific journals, reports, and presentations.
4. Customization: It offers extensive customization options to control every aspect of a plot, including colors, styles, labels, axes, and annotations, allowing users to tailor visualizations to their specific needs.
5. Integration: Matplotlib can be integrated seamlessly with other Python libraries and tools such as NumPy, Pandas, and Jupyter Notebooks, making it a versatile choice for data analysis and visualization workflows.

Some of the common plots that can be created using the Pyplot module of Matplotlib include:

1. Line Plot: Used to visualize data points connected by straight line segments, commonly used to show trends over time or continuous data.
2. Scatter Plot: Displays individual data points as markers on a two-dimensional plane, useful for visualizing relationships between two variables.
3. Histogram: Represents the distribution of a dataset by dividing it into bins and displaying the frequency of data points within each bin.
4. Bar Plot: Displays categorical data with rectangular bars whose lengths or heights represent the values of the corresponding categories.
5. Pie Chart: Represents data as sectors of a circle, with each sector's angle proportional to the relative frequency or percentage of the corresponding category.

A scatter plot is a type of plot that displays individual data points as markers on a two-dimensional plane. Each data point has a corresponding position determined by its values on the x-axis and y-axis. Scatter plots are commonly used to visualize relationships between two variables and identify patterns or trends in the data.

```python
import numpy as np
import matplotlib.pyplot as plt

# Set random seed for reproducibility
np.random.seed(3)

# Generate data for x and y
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))

# Plot scatter plot
plt.scatter(x, y)

# Add title, xlabel, ylabel
plt.title('Scatter Plot')
plt.xlabel('X')
plt.ylabel('Y')

# Show plot
plt.show()
```

The 'subplot()' function in Matplotlib is used to create multiple plots within the same figure. It allows you to arrange plots in a grid-like layout, specifying the number of rows and columns, as well as the position of each plot within the grid.

```python
import matplotlib.pyplot as plt
import numpy as np

# Data for line plots
x1 = np.array([0,1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])

x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])

x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])

x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])

# Create a figure and subplots
fig, axs = plt.subplots(2, 2)

# Plot line plots
axs[0, 0].plot(x1, y1)
axs[0, 0].set_title('Line 1')

axs[0, 1].plot(x2, y2)
axs[0, 1].set_title('Line 2')

axs[1, 0].plot(x3, y3)
axs[1, 0].set_title('Line 3')

axs[1, 1].plot(x4, y4)
axs[1, 1].set_title('Line 4')

# Adjust layout
plt.tight_layout()

# Show plots
plt.show()
```

A bar plot is a type of plot that presents categorical data with rectangular bars whose lengths or heights represent the values of the corresponding categories. It is commonly used to compare the values of different categories or to show the distribution of a categorical variable.

Bar plots are used for various purposes, including:

1. Comparison: Bar plots allow for easy comparison of values between different categories, making it simple to identify trends or patterns in the data.
2. Visualization of Categorical Data: They are used to visualize the distribution or frequency of categorical variables, providing a clear representation of the data.
3. Presentation of Metrics: Bar plots are often used to present metrics or performance indicators for different categories or groups, making it easier to understand and interpret the data.
4. Communication: They are used to communicate information effectively, especially when presenting data to a non-technical audience or stakeholders.

```python
import numpy as np
import matplotlib.pyplot as plt

# Data
company = np.array(['Apple', 'Microsoft', 'Google', 'AMD'])
profit = np.array([3000, 8000, 1000, 10000])

# Create a bar plot
plt.figure(figsize=(8, 6))
plt.bar(company, profit, color='skyblue')
plt.xlabel('Company')
plt.ylabel('Profit (in millions)')
plt.title('Profit of Companies')
plt.xticks(rotation=45)  # Rotate x-axis labels for better visibility
plt.grid(axis='y')  # Add gridlines for y-axis
plt.show()

# Create a horizontal bar plot
plt.figure(figsize=(8, 6))
plt.barh(company, profit, color='lightgreen')
plt.xlabel('Profit (in millions)')
plt.ylabel('Company')
plt.title('Profit of Companies (Horizontal)')
plt.grid(axis='x')  # Add gridlines for x-axis
plt.show()
```

A box plot, also known as a box-and-whisker plot, is a statistical visualization technique used to summarize the distribution of a continuous variable through five summary statistics: minimum, first quartile (Q1), median (Q2), third quartile (Q3), and maximum. It provides a graphical representation of the central tendency, variability, and presence of outliers in the data.

Box plots are commonly used for the following purposes:

1. Identifying Central Tendency and Variability: Box plots help visualize the spread of the data and identify the central tendency (median) and the spread (interquartile range) around the median.
2. Detecting Outliers: They provide a visual indication of potential outliers in the dataset, which are data points that significantly deviate from the overall distribution.
3. Comparing Distributions: Box plots are useful for comparing the distributions of multiple groups or datasets, allowing for easy visual comparison of medians, spreads, and presence of outliers.

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate random data for box plots
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

# Combine data into a list
data = [box1, box2]

# Create a box plot
plt.figure(figsize=(8, 6))
plt.boxplot(data, labels=['Box 1', 'Box 2'])
plt.title('Box Plot of Data')
plt.xlabel('Box')
plt.ylabel('Value')
plt.grid(True)
plt.show()
```
